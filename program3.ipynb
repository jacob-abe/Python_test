```
import seaborn as sns
import pandas as pd
import numpy as np
iris = sns.load_dataset('iris')
iris.head()


#Create a DataFrame
df = pd.DataFrame(iris)
print(df)
print('DF sum')
print(df.sum())
print('DF Sum axis=1')
print(df.sum(1))
print('DF mean')
print(df.mean())
print('DF std deviation')
print(df.std())
print('DF mode')
print(df.mode())
print('DF median')
print(df.median())

```
